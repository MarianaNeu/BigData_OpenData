# 1. Introducción a los datos abiertos

### ¿Qué son los datos abiertos?
> Los datos abiertos, tambien conocidos normalmente como "open data", son todos los datos que estan disponibles de forma pública, en un formato que puede ser legible por maquinas y aquellos datos que se pueden utilizar, reutilizar y redistribuir de manera libre y gratuita, sin ningun problema.
Estos datos pueden estar en cualquier formato, como hojas de cálculo, imágenes, grabaciones de audio o tambien archivos de texto, y pueden usarse para diversos fines, como por ejemplo para la investigación, el análisis y la toma de decisiones.

> Se promueve el uso de datos abiertos como un recurso público, con el objetivo de aumentar la transparencia, la responsabilidad y la innovación. El uso de los datos abiertos tambien permite la colaboración y el intercambio de conocimientos entre diferentes industrias, sectores y comunidades. Los datos abiertos pueden provenir de varias fuentes, como agencias gubernamentales, organizaciones internacionales, organizaciones sin fines de lucro y empresas privadas. Para que los datos sean abiertos, deben estar publicados en un formato estandarizado y estar disponibles a través de un portal público o API.

### Los Datos Abiertos en Paraguay
> En los ultimos años en Paraguay, se han implementado políticas y plataformas como el Portal de Datos Abiertos para fomentar la transparencia de los datos.Hay muchísimas razones por las cuales paraguay debe poseer datos abiertos, como por ejemplo:

> **- Transparencia y responsabilidad:** los ciudadanos pueden acceder a información sobre las acciones del gobierno y otras organizaciones públicas si es que los datos están abiertos. Esto es sumamente importante ya que aumenta la transparencia en la gestión pública y tambien permite la fiscalización del uso de los recursos públicos.

> **- Participación ciudadana:** Al proporcionar datos abiertos, se fomenta la participación de todos los ciudadanos y permite que se cree un ambiente favorable para que la sociedad participe en las decisiones tomadas. La creación de herramientas y aplicaciones que permiten a los ciudadanos interactuar con la información y contribuir a la mejora de los servicios públicos se puede lograr gracias a la utilización de datos abiertos.

> **- Innovación y desarrollo económico:** los emprendedores, las empresas y todos los desarrolladores de paraguay pueden utilizar los datos abiertos para crear nuevos bienes, servicios y soluciones tecnológicas. Esto impulsa muchisimo la creatividad general y tambien tiene el potencial de impulsar el crecimiento económico del país.

> **- Mejorar la eficiencia del gobierno:** al hacer que los datos estén disponibles de manera abierta, se promueve la eficiencia y la responsabilidad en la gestión del gobierno. Los datos abiertos pueden ayudar en la optimización de procesos, la toma de decisiones basadas en evidencia y la identificación de áreas de mejora.

> En Paraguay existe una ley llamada, **Ley "De Acceso a la Información Pública"**, la cual proporciona el marco legal para asegurar el acceso a la información pública. Esta ley fomenta la transparencia de las instituciones públicas y el acceso a la información. Además, la Política de Datos Abiertos fue establecida por la Dirección Nacional de Contrataciones Públicas (DNCP) para fomentar la apertura y reutilización de datos en el sector público.

> El Portal de Datos Abiertos fue creado por el Gobierno paraguayo para facilitar la divulgación y el acceso a los datos públicos del país. El portal ofrece una amplia gama de conjuntos de datos en diversos campos, incluida la educación, la salud, la economía, el medio ambiente y el transporte.


> El Catálogo de Datos Abiertos del Gobierno de Paraguay, ubicado en www.datos.gov.py, es el portal oficial de datos abiertos del estado paraguayo. Sirve como el principal punto de acceso a los datos abiertos que las instituciones gubernamentales han proporcionado de acuerdo con estándares y formatos abiertos. En este catálogo se incluyen conjuntos de datos (datasets) creados por instituciones, clasificados por categorías y presentados en formatos variados para permitir su descarga gratuita. Además, se incluyen licencias gratuitas para el uso, reutilización y distribución de cada dataset.

> Cada vez más, Paraguay ha demostrado su compromiso con la apertura de datos. Ha creado un Plan de Acción de Datos Abiertos para fomentar la apertura y el uso de datos en el país y participa en iniciativas internacionales como la Alianza para el Gobierno Abierto (OGP).
En Paraguay, los datos abiertos se han utilizado para una variedad de propósitos, incluida la visualización de datos, la investigación académica, la creación de aplicaciones móviles y la generación de conocimiento.
````python
def cleanData(text):
    text = re.sub('[^\w\s]', '', text)  # Elimina los caracteres especiales
    text = re.sub(r'[^\x00-\x7F]+', '', text)  # Elimina los caracteres no ASCII (como los emojis)
    text = ''.join(c for c in unicodedata.normalize('NFD', text) if unicodedata.category(c) != 'Mn')  # Elimina las tildes
    return text
````

````python
import unicodedata

def remove_accents_and_degrees(df):
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column] = df[column].apply(lambda x: unicodedata.normalize('NFKD', str(x)).encode('ASCII', 'ignore').decode('utf-8'))
    return df
````

> De acuerdo a los resultados obtenidos, no ocurre nada atípico con las transferencias y todo lo transferido es depositado. Si esto no salia asi podria significar que el gobierno no estaba depositando el monto total de la transferencia de recursos, pero por suerte ese no es el caso.

````python
import pandas as pd

# Calcular la diferencia entre las fechas de depósito y aprobación
df['diferencia_anhos'] = df['anho_deposito'] - df['anho_aprobacion']
df['diferencia_meses'] = df['mes_deposito'] - df['mes_aprobacion']
df['diferencia_dias'] = df['dia_deposito'] - df['dia_aprobacion']

# Convertir los valores negativos a positivos y ajustar la diferencia en meses y días
df['diferencia_meses'] += df['diferencia_anhos'] * 12

# Ajustar los días para que se sumen a los meses
df['diferencia_meses'] += df['diferencia_dias'] // 30
df['diferencia_dias'] = df['diferencia_dias'] % 30

# Calcular el promedio de la diferencia en años, meses y días
promedio_meses = df['diferencia_meses'].mean()
promedio_anhos = promedio_meses // 12
promedio_meses = promedio_meses % 12
promedio_dias = df['diferencia_dias'].mean()

# Imprimir el resultado en el formato deseado
print(f"Promedio: {promedio_anhos:.0f} años, {promedio_meses:.0f} meses y {promedio_dias:.0f} días")
````

Se puede concluir que en promedio el tiempo desde que la transferencia fue aprobada hasta la realizacion de la transferencia no es mucho, menos de 3 meses. La mayoria de las transferencias fueron aprobadas a finales del 2017 y luego fueron realizadas a comienzos de 2018.
